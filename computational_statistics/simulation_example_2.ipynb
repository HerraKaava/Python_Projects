```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vuosittain syntyvien lasten lkm
n = 50000
```

```python
# Simuloidaan syntymien lukumäärä kullekin vuoden päivälle olettaen sama 
# todennäköisyys jokaiselle päivälle
x = np.random.multinomial(n=n, pvals=[1/365]*365, size=1)
```

```python
x
```

```python
# Vuoden maksimi samana päivänä syntyneitä
np.max(x)
```

```python
# Toistetaan simulaatio 10000 kertaa
def max_per_day():
    return np.max(np.random.multinomial(n=n, pvals=[1/365]*365, size=1))

def gen_simulations(reps):
    return [max_per_day() for _ in range(reps)]
```

```python
v = gen_simulations(reps=10000)
```

```python
plt.hist(v, edgecolor='black', bins=20);
```

```python
# Saman voi tehdä tehokkaammin antamalla simulointien lkm suoraan numpyn funktiolle
gen_sims = np.random.multinomial(n=n, pvals=[1/365]*365, size=10000)
```

```python
gen_sims.shape
```

```python
# Maksimi riveittäin
row_max = np.max(gen_sims.T, axis=0)
```

- Rivit sisältävät vuodet, ja vuosia simuloitiin 10000.
- Maksimi riveittäin siis palauttaa jokaiselta vuodelta maksimin samana päivänä syntyneitä lapsia.

```python
plt.hist(row_max, bins=40, edgecolor='black');
```
