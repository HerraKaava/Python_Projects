```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p_X(x):
    return np.maximum(0, 1 - np.abs(1 - x))
```

```python
x = np.linspace(0, 2, 1000)
pdf = p_X(x)
plt.figure(figsize=(8,6))
plt.plot(x, pdf);
```

```python
def rejection_sampling(n):
    """
    Args:
        n: the number of samples to generate
    """
    res = np.zeros(n)
    i = 0
    while i < n:
        # Tuotetaan tasajakautuneita pisteitä suorakaiteesta [0,2]x[0,1]
        x = np.random.uniform(low=0.0, high=2.0, size=1)
        y = np.random.uniform(low=0.0, high=1.0, size=1)
        
        # Hyväksytään pisteet, jotka osuvat tiheysfunktion käyrän alapuolelle
        if y <= p_X(x):
            res[i] = x.item()
            i += 1
    return res
```

```python
samples = rejection_sampling(n=10000)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, pdf, color='red', zorder=2)
plt.hist(samples, density=True, bins=50, edgecolor='black', zorder=1);
```
