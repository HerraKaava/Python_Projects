```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_random_pairs(n):
    """
    Simulates n uniformly distributed random pairs of numbers from the set [-1,1]^2
    """
    x = np.random.uniform(low=-1, high=1, size=n)
    y = np.random.uniform(low=-1, high=1, size=n)
    k = x**2 + y**2 <= 1**2
    plt.scatter(x, y, c=np.where(k, 'blue', 'green'))
    plt.gca().set_aspect('equal')    # Set the aspect ratio to 1:1
    return k
```

```python
n = 100000
k = simulate_random_pairs(n=n)
```

```python
# Pi approximation
4 * np.sum(k) / n
```

Explanation on why the formula above works in estimating $\, \pi. \,$

We are generating $\, n \,$ random points in the square $\, [-1,1] \times [-1,1]. \,$ These random points are uniformly distributed $\,(x \sim \text{Uniform}(0,1), \, y \sim \text{Uniform}(0,1)), \,$ which means that the points are equally likely to land anywhere in the square. Let $\, k \,$ be the number of points that land inside the unit circle. Then $\, \frac{k}{n} \,$ is the *fraction* of points that land inside the unit circle. Imagine throwing darts randomly at a square dartboard with a circle drawn inside it. **If the darts are thrown *uniformly*, the chance that a dart hits the circle is proportional to the area of the circle.** For example, if 25% of the darts land in the circle (as $\, n := \text{"the number of the throws"} \,$ grows), then the circle must be 25% of the square's area. Hence

$$ \frac{k}{n} \approx \mathbb{P}(\text{A point lands in the circle}) = \frac{\text{Area of the circle}}{\text{Area of the square}} = \frac{\pi r^2}{s^2}, $$

where $\, r \,$ is the radius of the circle, and $\, s \,$ is the length of one side of the square. In our case, $\, r=1 \,$ and $\, s=2, \,$ so

$$ \frac{k}{n} \approx \mathbb{P}(\text{A point lands in the circle}) = \frac{\text{Area of the circle}}{\text{Area of the square}} = \frac{\pi r^2}{s^2} = \frac{\pi}{4}. $$ Solving for $\, \pi \,$ we get

$$ \pi \approx 4 \cdot \frac{k}{n} $$
