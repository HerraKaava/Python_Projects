<h2 align="left">Understanding the Figure Object</h2>

- The more comprehensive Matplotlib OOP API (Object Oriented Prograrmming Application Programming Interface) makes use of a **Figure object**.
- We basically create a general figure object as a blank canvas, and then we add axes to this figure object, and plot on those set of axes.

```python
plt.figure();
```

- When we call plt.figure(), we create a figure object (which is a blank canvas because we haven't added axes to it), which has a default size of 432x288.
- Note that the the figure object is not visible until we add the axes to it.
- If we wanted to adjust the pixel size, we could specify inside the plt.figure() call an argument figsize=(n, k).

```python
plt.figure(figsize=(10, 10));
```

This 'blank canvas' can be saved to a variable, and then this variable can be used to add elements to the figure.

```python
# Create the 'blank canvas'
fig = plt.figure()

# Add axes to the blank canvas
axes = fig.add_axes([0, 0, 1, 1])
```

**add_axes()**
- The first two arguments adjusts the lower left corner of the axes.
- The third and fourth arguments adjusts the width and height of the axes (in that order).

```python
# Create the blank canvas,
# and adjust figure size and DPI (dots per inch --> i.e. the sharpness of the figure)
figz = plt.figure(figsize=(4, 3), dpi=200)

# Notice that one does not need to save the axes object into a variable,
# if one does not need to manipulate the axes any further.
figz.add_axes([0, 0, 2, 1]);
```

- Notice how the figure is much wider than the one before because I adjusted the width of the figure with the third argument.
- Keep in mind that increasing the DPI also indirectly affects the figure size because it acts like a magnifying glass.

Next we'll take a look at how to add **multiple set of axes** to a **single figure object**.

```python
fig_multiple = plt.figure(dpi=100)

# LARGE PLOT

axes_ena = fig_multiple.add_axes([0, 0, 1, 1])
axes_ena.plot(a, b)

axes_ena.set_xlim(0, 8)
axes_ena.set_ylim(0, 8000)
axes_ena.set_xlabel("a", size=15)
axes_ena.set_ylabel("b", rotation=0, labelpad=15, size=15)
axes_ena.set_title("Power of 4", size=15)

# SMALL PLOT

axes_dio = fig_multiple.add_axes([0.2, 0.5, 0.25, 0.25])
axes_dio.plot(a, b)

axes_dio.set_xlim(1, 2)
axes_dio.set_ylim(0, 50)
axes_dio.set_xlabel("a", size=12)
axes_dio.set_ylabel("b", rotation=0, labelpad=10, size=12)
axes_dio.set_title("Zoomed IN", size=12);
```

- I encourage to play around with the add_axes() function's arguments to get an idea how the size, scale, and location of the axes changes when you change the arguments.

- This type of plot can be used in data analytics to have a zoomed in section of your plot as an insert.
- So you can use this type of plotting to insert a zoomed in section of the actual bigger plot.

- When you want to add labels, limits, etc to the plots seperately, you just have to call the relevant methods on the **specific axes** you wish to add them on.

- Note that you can add as many axes to a single figure object as you want.
- You can also technically place them where ever you want with the arguments of the add_axes() method.

<h2 align="left">Subplots Functionality</h2>

- The **plt.subplots()** call allows one to easily create a figure and axes objects in a side by side formations.
- The plt.subplots() command returns **a tuple** containing the **figure canvas** and a **numpy array** holding the **axes objects**.


- An important note regarding subplots is that all the arguments learned previously regarding the figure object can also be applied to the plt.subplots() function.
- When pressing shift + tab on the plt.subplots() function, it comes apparent that it includes a **fig_kw** parameter, indicating that it accepts all keywords associated with the figure object.


- plt.subplots returns a tuple, which is commonly labeled as **fig, axes** by convention.

```python
a = np.linspace(0, 10, 11)
```

```python
b = a ** 4
```

```python
x = np.arange(0, 10)
```

```python
y = 2*x
```

Note that when plt.subplots() is called, a tuple containing the figure canvas and the axes objects is returned, and they need to be unpacked into **two separate variables.**

As for the arguments, we have four main arguments that we can modify:
- **nrows** (how many rows you want your subplots to have)
- **ncols** (how many columns you want your subplots to have)
- **sharex** (if you want all / some of your subplots to share the x-axis. Note that this is False by default)
- **sharey** (if you want all / some of your subplots to share the y-axis. Note that this is False by default)

```python
fig4, axes4 = plt.subplots(nrows=1, ncols=1);
```

```python
type(axes4)
```

```python
# A subplot with one row and two columns
fig5, axes5 = plt.subplots(nrows=1, ncols=2);
```

```python
type(axes5)
```

- Note that as soon as we start asking for more than one row and one column (i.e. more than one graph), the type of the axes object changes from ***matplotlib.AxesSubplot*** to ***numpy.ndarray***.
- HOWEVER, the AxesSubplot objects didn't disappear anywhere. They just are now contained inside a numpy array.

Now, it is important to note that, since we now have a numpy array object, the **shape attribute** is available.

```python
axes5.shape
```

- We already know that we can **index** numpy array objects. The indexing works here too.

```python
axes5[0]
```

```python
axes5[1]
```

- It can be observed that, upon indexing the numpy array, the return values are AxesSubplot objects.
- **One can use this indexing to plot different plots on the different axes**.
- axes5[0] corresponds to the first plot, and axes5[1] corresponds to the second plot.
- The indexing here is no different from indexing a normal numpy array.
- For instance, if we had more rows and columns, then axes[2, 3] would correspond to the plot that is on third row and fourth column (indexing starts from zero).

```python
fig6, axes6 = plt.subplots(nrows=1, ncols=2)

axes6[0].plot(x, y)
axes6[1].plot(a, b);
```

- Two distinct plots are now present, as can be seen.
- Note that the **spacing** here is not that great. We'll learn how to edit this in later on.

Let's add more rows to the subplots.

```python
fig7, axes7 = plt.subplots(nrows=2, ncols=2);
```

```python
axes7.shape
```

```python
axes7
```

- As expected, with the addition of the second row, our axes object is now **two-dimensional**.

```python
fig8, axes8 = plt.subplots(nrows=2, ncols=2)

# Add a plot to axes in row 1, column 1 (indexing starts from zero).
axes8[0, 0].plot(a, b)

# Add a plot to axes in row 2, column 2.
axes8[1, 1].plot(x, y);
```

One can also use a loop to plot on the axes.

```python
fig9, axes9 = plt.subplots(nrows=3, ncols=1)

for ax in axes9:
    ax.plot(x, y)
```

Next, let's cover the **spacing issue**.

**Automatic Spacing**

- For automatic spacing, after we have created all of the plots, we can call plt.tight_layout() to deal with the spacing issue.

Let's re-create the axes above and use automatic spacing.

```python
fig10, axes10 = plt.subplots(nrows=3, ncols=1)

for ax in axes10:
    ax.plot(x, y)

# Automatic spacing
plt.tight_layout()
```

- The plt.tight_layout() call **automatically** attempts to make sure that no set of axes tickers are overlapping.

**Manual Spacing**

For manual spacing, one can to call the **subplots_adjust()** function on the **fig object**.
- See the link below for an explanation of the parameters of the subplots_adjust() function.
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

```python
fig11, axes11 = plt.subplots(nrows=2, ncols=2)

axes11[0, 0].plot(x, y)
axes11[0, 1].plot(a, b)
axes11[1, 0].plot(x, y)
axes11[1, 1].plot(a, b)

# Manual spacing
fig11.subplots_adjust(wspace=0.4, hspace=0.4);
```

- The wspace and hspace parameters accept float values between 0 and 1 as an argument.
- The previously mentioned arguments represents the fraction of the average axis width / height.
- For instance, if we set wspace = 1, then between the subplots we can fit one subplot in terms of width. 
- For example, setting wspace=1 allows one subplot's width of space between the subplots.
- Same holds true for hspace=1.

This can be illustrated by actually creating the previously mentioned scenario.

```python
fig12, axes12 = plt.subplots(nrows=2, ncols=2)

axes12[0, 0].plot(x, y)
axes12[0, 1].plot(a, b)
axes12[1, 0].plot(x, y)
axes12[1, 1].plot(a, b)

fig12.subplots_adjust(wspace=1, hspace=1);
```

- As seen above, now we could fit both vertically and horizontally one subplot between the axes.
- To clarify, the spacing becomes bigger as the wspace and hspace arguments get closer to 1.
- To clarify, as the 'wspace' and 'hspace' arguments approach 1, the spacing increases.

One last **important note** regarding the subplots functionality is that the rest of the parameters (e.g. title, xlabel, ylabel, xlim, ylim) work in the same manner as the plotting, meaning that one can refer to them by their array index position.

```python
fig13, axes13 = plt.subplots(nrows=2, ncols=2, dpi=80)

axes13[0, 0].plot(x, y)
axes13[0, 1].plot(a, b)
axes13[1, 0].plot(x, y)
axes13[1, 1].plot(a, b)

fig13.suptitle("Figure Level Title", fontsize=13)

axes13[0, 0].set_title("First Plot")
axes13[0, 1].set_title("Second Plot")
axes13[1, 0].set_title("Third Plot")
axes13[1, 1].set_title("Fourth Plot")

axes13[0, 0].set_xlabel("First")
axes13[0, 1].set_xlabel("Second")
axes13[1, 0].set_xlabel("Third")
axes13[1, 1].set_xlabel("Fourth")

axes13[0, 0].set_ylabel("y1", rotation=0, labelpad=10)
axes13[0, 1].set_ylabel("y2", rotation=0, labelpad=10)
axes13[1, 0].set_ylabel("y3", rotation=0, labelpad=10)
axes13[1, 1].set_ylabel("y4", rotation=0, labelpad=10)

fig13.subplots_adjust(wspace=0.6, hspace=0.8);
```

<h3 align="left">Legends</h3>

```python
a = np.linspace(0, 10, 11)
```

```python
fig14 = plt.figure()

ax14 = fig14.add_axes([0, 0, 1, 1])

ax14.plot(a, a, label="a vs. a")

ax14.plot(a, a**2, label="a vs. a squared")

ax14.legend(loc=0);
```

- Note that the legend() function has a "loc" (location) parameter.
- We can specify the location that we want the legend to be placed on.
- By default loc=0, which means "best location" as in matplotlib automatically tries to fit it to the best location so that it won't interfere with the displayed graph (see the table below for the complete list of all the intergers that the loc parameter accepts, and the corresponding locations)
- Note that we could also just pass in the location as a string, e.g. loc="upper right".
- If we want to manually set the location, we can pass a tuple of integers to the loc parameter as an argument.
- The logic in the tuple is the same as the logic in the subplots_adjust() function's wspace and hspace arguments.

```python
from tabulate import tabulate

data = [
    ['Location String', 'Location Code'],
    ['best', 0],
    ['upper right', 1],
    ['upper left', 2],
    ['lower left', 3],
    ['lower right', 4],
    ['right', 5],
    ['center left', 6],
    ['center right', 7],
    ['lower center', 8],
    ['upper center', 9],
    ['center', 10]
]

print(tabulate(data))
```

<h3 align="left">Styles</h3>

- The plot() methods's **lw** (linewidth) parameter can be used to modify the width of the curve.
- lw accepts integer / float as an argument, which represents a ratio. 

$ \quad \quad \boldsymbol{\rightarrow} \,$ The default value is 1, so for instance, setting lw=10 makes the curve 10 times wider than the default curve.

```python
fig16 = plt.figure()

ax16 = fig16.add_axes([0, 0, 1, 1])

ax16.plot(a, a, color="red", lw=10)

ax16.plot(a, a+1, color="green", lw=5);
```

Another useful plot() methods's parameter is **ls** (linestyle), which can be used to modify the linestyle of the curve.

The possible linestyle options are
- "**--**" 
- "**–**"
- "**-.**" 
- "**:**" 

```python
fig17 = plt.figure(dpi=100)
ax17 = fig17.add_axes([0, 0, 1, 1])

ax17.plot(a, a, color="red", lw=1, ls="--")

ax17.plot(a, a+1, color="blue", lw=1, ls="-")

ax17.plot(a, a+2, color="black", lw=1, ls="-.")

ax17.plot(a, a+3, color="green", lw=1, ls=":");
```

Although the plot may appear to depict our data points as continuous, upon examining the values within the variable **a** it becomes evident that **a** is, in fact, a discrete variable rather than a continuous variable.

```python
a
```

If we want to see where the actual points (values) are in the curve, we can use the plot() method's **marker** parameter. 
- To see all the possible marker styles available, click [HERE](https://matplotlib.org/stable/api/markers_api.html)

```python
fig18 = plt.figure()
ax18 = fig18.add_axes([0, 0, 1, 1])

ax18.plot(a, a, color="red", lw=1, ls="--", marker="D")

ax18.plot(a, a**2, color="blue", lw=1, ls=":", marker="*", markersize=10)

ax18.grid(True);
```

- It is worth noting that the 'markersize' parameter is also available for adjusting the size of the markers, as the name implies.
- **ax.grid(True)** adds grids to the background for better readability.

For more styling and visualizations, see the Matplotlib gallery [HERE](https://matplotlib.org/stable/).
