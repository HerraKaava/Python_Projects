```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def L(p, N=10, D=4):
    return p**D * (1-p)**(N-D)
```

```python
p_vals = np.linspace(0.0, 1.0, 1000)
```

```python
likelihoods = np.array([L(p=i) for i in p_vals])
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(p_vals, likelihoods, zorder=2)
ax.set_xlabel('Risk')
ax.set_ylabel('Likelihood')
ax.grid(True, alpha=0.3, zorder=1)
```

- It is visible from the plot that the maximum likelihood is achieved with $\, \pi \, \text{(risk)} = 0.4. \,$
- This (the maximum likelihood estimate of $\pi$) could be easily solved analytically as well via maximum likelihood estimation.
